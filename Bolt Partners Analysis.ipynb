```python
'''STEP 1: Exploratory Data Analysis | Preparation & Cleaning(you may skip to Step 2, if you wish)'''
import pandas as pd
import numpy as np
#load the data and explore it
url="https://drive.google.com/uc?id=1lxErXic4KgHx6XcusuJ8ufABn8f0ZbXM&export=download"
df=pd.read_csv(url)

'''check the hypothesis H0:every transaction has deposit for a container'''
df.sort_values(by='Order ID', ascending=False)
#display(df)

'''let's check whether a randomly chosen order w/o deposit is the only with such ID (whether it has deposit record elsewhere in the dataset)'''
rand_order=df.query("`Order ID` ==  'ZY4'")
#display(rand_order)
'''as we can see, ZY4 is the only order w/such ID, so we count number of orders and deposits to check how many more orders are missing the deposit(or its record)'''
prod_btl_count=df[df['Title']=='ProductBottle'].shape[0]
dep_count=df[df['Title']=='Deposit'].shape[0]
#display("Number of transactions: ", prod_btl_count)
#display("Number of deposits:", dep_count)

'''there is clearly smth wrong, so we have to reject H0 in favour of alternative hypothesis H1:data is missing, not every order has a deposit(or record of it)'''
'''good news deposit would typically be an upfront payment, doesn't affect net revenue,
so we can disregard corresponding rows'''
df2=df[df['Title']!='Deposit']
#display(df2)

'''STEP 2: Exploratory Data Analysis | Summary Statistics:Financial Performance & Top Partners'''
net_rev=df2['Total (without VAT)'].sum()-df2['Bolt Service Fee'].sum()
display("Net Revenue: ", net_rev)
num_orders=df2['Order ID'].nunique()
display("Number of Orders: ", num_orders)
ave_order_val=df2['Total (without VAT)'].sum()/num_orders
display("Average Order Value: ", ave_order_val)
df2['Order Created Time']=pd.to_datetime(df['Order Created Time'])
df2['Order Date']=df2['Order Created Time'].dt.to_period('D')
df2['Net Revenue']=df2['Total (without VAT)']-df2['Bolt Service Fee']
top_sell_orders = df2.groupby(['Order ID','Order Date'])['Net Revenue'].sum().reset_index().sort_values(by='Net Revenue',ascending=False)
display("Top Selling Orders: ", top_sell_orders.head(10))

'''let's compare revenue for 2021 and 2022'''
df_2022=df2[df2['Order Date']>'2021-12-31']
df_2022['Net Revenue']=df_2022['Total (without VAT)']-df_2022['Bolt Service Fee']
top_sell_orders_2022 = df_2022.groupby(['Order ID','Order Date'])['Net Revenue'].sum().reset_index().sort_values(by='Net Revenue',ascending=False)
display("Top Selling Orders 2022: ", top_sell_orders_2022.head(10))

'''let's compare average revenue of 2022 and 2021'''
df_2021=df2[df2['Order Date']<='2021-12-31']
net_rev_2021=df_2021['Total (without VAT)'].sum()-df_2021['Bolt Service Fee'].sum()
display("Net Revenue 2021: ", net_rev_2021)
num_orders_2021=df_2021['Order ID'].nunique()
ave_order_val_2021=df_2021['Total (without VAT)'].sum()/num_orders_2021
display("Average Order Value 2021: ", ave_order_val_2021)
net_rev_2022=df_2022['Total (without VAT)'].sum()-df_2022['Bolt Service Fee'].sum()
display("Net Revenue 2022: ", net_rev_2022)
num_orders_2022=df_2022['Order ID'].nunique()
ave_order_val_2022=df_2022['Total (without VAT)'].sum()/num_orders_2022
display("Average Order Value 2022: ", ave_order_val_2022)
'''2021 seemes to be more profitable, than 2022, because 2021 is Dec, but 2022 is Jan!'''
'''Now, let's see the most profitable restaurants'''
top_net_rev_partners = df2.groupby(['Restaurant ID'])['Net Revenue'].sum().reset_index().sort_values(by='Net Revenue',ascending=False)
display(top_net_rev_partners)
```

```python
most_profit_partners = df2.groupby(['Restaurant ID'])['Bolt Service Fee'].sum().reset_index().sort_values(by='Bolt Service Fee',ascending=False)
display(most_profit_partners)
number_orders = df2.groupby('Restaurant ID')['Order ID'].count().reset_index()
number_orders = number_orders.rename(columns={'Order ID': 'Number_of_Orders'})
number_orders = number_orders.sort_values(by='Number_of_Orders', ascending=False)
display(number_orders)
peak_periods = df2.groupby('Order Date')['Order ID'].count().reset_index().rename(columns={'Order ID': 'Number_of_Orders'}).sort_values(by='Number_of_Orders', ascending=False)
display(peak_periods)
```
